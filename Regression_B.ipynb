```python
# Linear regression

# Necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import  statsmodels.formula.api as smf
from scipy import stats
from sklearn.metrics import mean_squared_error
```

```python
# load the data file
df=pd.read_excel("/content/drive/My Drive/python_data/RegressionData/PrawnTrawling.xls")
df.head()
```

```python
# rename some columns
df=df.rename(columns={'Score 1':'Score_1',
                   'Score 2':'Score_2',})
df.head()
```

```python
# Build a multiple regression model with Score_1
regPrawnT_S1 = smf.ols('Score_1 ~ Latitude + Longitude + Depth', data=df).fit()
print(regPrawnT_S1.summary())
```

```python
# Build a multiple regression model with Score_2
regPrawnT_S2 = smf.ols('Score_2 ~ Latitude + Longitude + Depth', data=df).fit()
print(regPrawnT_S2.summary())
```

```python
# load the data file
df=pd.read_excel("/content/drive/My Drive/python_data/RegressionData/mileage_2_xls.xls")
df.head()
```

```python
# Build a single regression model for MPG and HP
regMPG_HP = smf.ols('MPG ~ HP', data=df).fit()
print(regMPG_HP.summary())
```

```python
# Build a single regression model for MPG and inverse of HP
regMPG_HP_Inv = smf.ols('MPG ~ HP_Inv', data=df).fit()
print(regMPG_HP_Inv.summary())
```

```python
# Build a multiple regression model for MPG along with HP and WT
regMPG_HP_WT = smf.ols('MPG ~ HP + WT', data=df).fit()
print(regMPG_HP_WT.summary())
```

Look suddenly HP becomes insignifacnt with a P value more than 0.05. This happens because of multi-collinearity between HP and WT. Try to do scatter plot to see the corelation

```python
# see the scatter plot
plt.scatter(df.HP,df.WT) 
plt.xlabel("HP")
plt.ylabel("WT")
```

```python
# Build a multiple regression model for MPG along with WT
regMPG_WT = smf.ols('MPG ~ WT', data=df).fit()
print(regMPG_WT.summary())
```

```python
# Build a multiple regression model for MPG along with log transformation (base of 10) of WT
regMPG_log_WT = smf.ols('MPG ~ Log_WT', data=df).fit()
print(regMPG_log_WT.summary())
```

```python
# log transformation of independent variable WT to the base 'e'
regMPG_logE_WT =smf.ols('MPG ~ np.log(WT)', data=df).fit()
print(regMPG_logE_WT.summary())
```
